```python
import pandas as pd
from sklearn.cluster import KMeans
from scipy import sparse
import os
# クラスター分類数
n = 20

DATASET_FILE_PATH = "datasets/movie_lens/ml-latest-small/ratings.csv"
MOVIE_NAME_FPATH  = "datasets/movie_lens/ml-latest-small/movies.csv"
```

```python
%matplotlib inline
dataset_file_path = os.path.normpath(os.path.join(os.path.abspath(""), "../../" , DATASET_FILE_PATH))
movie_file_path   =  os.path.normpath(os.path.join(os.path.abspath(""), "../../" , MOVIE_NAME_FPATH))
```

```python
df = pd.read_csv(dataset_file_path)
max_userId  = df.max().ix['userId'].astype(int)
max_movieId = df.max().ix['movieId'].astype(int)
```

```python
# 映画名の読み込み
movies = {}
movies_df = pd.read_csv(movie_file_path)
for i, row in movies_df.iterrows():
    movies[row["movieId"]] = row["title"]

print(movies[1])
print("映画名の読み込み完了")
```

```python
# 疎行列への設定
mat = sparse.lil_matrix((max_userId, max_movieId))
for i, row in df.iterrows():
    mat[row["userId"] - 1, row["movieId"]  - 1] = row["rating"]

print("疎行列処理完了")
```

```python
# クラスタリング
kmeans = KMeans(n_clusters=n)
kmeans.fit(mat)
```

```python
clusters = [[] for _ in range(n)]
for i, label in enumerate(kmeans.labels_):
    if ((i+1) in movies) == False :
        movies[i+1] = i+1

    clusters[label].append(movies[i + 1])

print("クラスタリング処理完了")
```

```python
# クラスタのサイズ順（昇順）に表示
clusters = sorted(clusters, key=len)
for i in range(n):
    print("-" * 60)
    for m in clusters[i]:
        print(m)
```
